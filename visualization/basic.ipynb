```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
np.random.seed(1000)
# ndarray 객체를 plot의 인수로 넘기면 자동으로 x를 index로 처리
y = np.random.standard_normal(20)
```

```python
x = range(len(y))
# plt.plot(x, y)
plt.plot(y.cumsum())
plt.grid(True) # 그리드 추가
# plt.axis('equal') # x, y scale을 동일하게 한다
# plt.axis('off') # 축과 라벨을 없앤다
# plt.axis('scaled') # 스케일이 갖도록 크기를 조정
```

```python
plt.plot(y.cumsum())
plt.grid(True)
# 직접 축 범위 설정
plt.xlim(-1, 20)
plt.ylim(np.min(y.cumsum()) - 1, np.max(y.cumsum()) + 1)
```

```python
plt.figure(figsize=(7, 4))
# figsize 인수를 사용하여 그림의 크기를 정의.
plt.plot(y.cumsum(), 'b', lw=1.5)
# r: red, +: 덧셈 모양
plt.plot(y.cumsum(), 'r+')
plt.grid(True)
plt.axis('tight')
```

```python
# 2차원 자료
# 다른 x, y축을 쓸 수도 있다.
# 하나는 선, 하나는 막대로 표현해야 할 수도 있

# example
np.random.seed(2000)
y = np.random.standard_normal((20, 2)).cumsum(axis=0)
```

```python
plt.figure(figsize=(7, 4))
# lw: line width
plt.plot(y, lw=1.5)
plt.plot(y, 'ro')
plt.grid(True)
plt.xlabel('index')
plt.ylabel('value')
plt.title('A Simple Plot')
```

```python
plt.figure(figsize=(7, 4))
# 분리해서 그리기
plt.plot(y[:,0], lw=1.5, label='1st')
plt.plot(y[:,1], lw=1.5, label='2st')
plt.plot(y, 'ro')
plt.grid(True)
# loc 0: 위치를 적당한 곳으로 자동 위치
plt.legend(loc=0)
plt.xlabel('index')
plt.ylabel('value')
plt.title('A Simple Plot')
```

```python
y[:, 0] = y[:, 0] * 100
plt.figure(figsize=(7, 4))
plt.plot(y[:,0], lw=1.5, label='1st')
plt.plot(y[:,1], lw=1.5, label='2st')
plt.plot(y, 'ro')
plt.grid(True)
plt.legend(loc=0)
plt.xlabel('index')
plt.ylabel('value')
plt.title('A Simple Plot')
# scale 차이로 하나의 y축만 써서 시각적으로 표현하기 힘듦
```

```python
# 해결법 1: 두개의 y축을 사용하는 방법
# 해결책 2: 두개의 subplot을 사용하는 방법

# 해결책 1
# 첫 번째 자료를 첫 번째 축(왼쪽)에 대응하여 그린다.
fig, ax1 = plt.subplots()
plt.plot(y[:, 0], 'b', lw=1.5, label='1st')
plt.plot(y[:, 0], 'ro')
plt.grid(True)
plt.legend(loc=8)
plt.xlabel('index')
plt.ylabel('value 1st')
plt.title('A Simple Plot')

# 두 번째 자료를 두 번째 축(오른쪽)에 대응하여 그린다.
ax2 = ax1.twinx()
plt.plot(y[:, 1], 'g', lw=1.5, label='2nd')
plt.plot(y[:, 1], 'ro')
plt.legend(loc=0)
plt.ylabel('value 2nd')
```

```python
# 해결책 2
# 두 subplot을 구분해서 그리기
plt.figure(figsize=(7, 5))
# subplot(211): (numrows, numcols, fignum) -> 총 가능한 fignum = numrows * numcols
plt.subplot(211)
plt.plot(y[:, 0], lw=1.5, label='1st')
plt.plot(y[:, 0], 'ro')
plt.grid(True)
plt.legend(loc=0)
plt.axis('tight')
plt.ylabel('value')
plt.title('A Simple Plot')
plt.subplot(212)
plt.plot(y[:, 1], 'g', lw=1.5, label='2nd')
plt.plot(y[:, 1], 'ro')
plt.grid(True)
plt.legend(loc=0)
plt.axis('tight')
plt.xlabel('index')
plt.ylabel('value')
```

```python
# 해결책 2
# 두 subplot을 구분해서 그리기
plt.figure(figsize=(9, 4))
plt.subplot(121)
plt.plot(y[:, 0], lw=1.5, label='1st')
plt.plot(y[:, 0], 'ro')
plt.grid(True)
plt.legend(loc=0)
plt.axis('tight')
plt.xlabel('index')
plt.ylabel('value')
plt.title('1st Data Set')
plt.subplot(122)
plt.bar(np.arange(len(y)), y[:, 1], width=0.5, color='g', label='2nd')
plt.grid(True)
plt.legend(loc=0)
plt.axis('tight')
plt.xlabel('index')
plt.ylabel('value')
```

```python
# 두 개의 금융 시계열 자료의 수익률을 동시에 볼 때!
y = np.random.standard_normal((1000, 2))
plt.figure(figsize=(7, 5))
plt.plot(y[:, 0], y[:, 1], 'ro')
plt.grid(True)
plt.xlabel('1st')
plt.ylabel('2nd')
plt.title('Scatter Plot')
```

```python
# scatter 함수를사용할 수도 있다.
plt.figure(figsize=(7, 5))
plt.scatter(y[:, 0], y[:, 1], marker='o')
plt.grid(True)
plt.xlabel('1st')
plt.ylabel('2nd')
plt.title('Scatter Plot')
```

```python
# plot, scatter 함수 차이 점
# 여러가지 색을 써서 3번째 차원을 추가 할 수 있음

# example
# 세번째 차원
c = np.random.randint(0, 10, len(y))
plt.figure(figsize=(7, 5))
plt.scatter(y[:, 0], y[:, 1], c=c, marker='o')
plt.colorbar()
plt.grid(True)
plt.xlabel('1st')
plt.ylabel('2nd')
plt.title('Scatter Plot')
```

```python
# 히스토그램
plt.figure(figsize=(7, 4))
# bins: 빈도 구분값의 수 -> 몇개의 histogram으로 보이게 할 것인가
print(y)
plt.hist(y, label=['1st', '2nd'], bins=10, histtype='barstacked')
plt.grid(True)
plt.legend(loc=0)
plt.xlabel('value')
plt.ylabel('frequency')
plt.title('Histogram')

# plt.hist 에 다영한 옵션을 잘 활용해보기
# cumulative, histtype: bar, barstacked, step, stepfilled
```
